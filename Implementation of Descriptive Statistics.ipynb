1.implementation of descriptive statistics

```python
import pandas as pd
import numpy as np
from scipy import stats

# Data
data = {
    'Age': [23, 25, 31, 35, 45, 55, 60, 65, 70, 72],
    'Income': [20000, 25000, 32000, 35000, 45000, 55000, 58000, 60000, 70000, 72000]
}

# Create DataFrame
df = pd.DataFrame(data)
print(df)

# Set float format
pd.options.display.float_format = '{:.2f}'.format

# Function for descriptive statistics
def descriptive_statistics(df):
    stats_summary = {}

    # Loop through each column
    for column in df.columns:
        # Calculate mode
        mode_result = stats.mode(df[column], keepdims=True)
        mode_value = mode_result.mode[0] if mode_result.count[0] > 0 else np.nan

        # Create statistics summary for each column
        stats_summary[column] = {
            'count': df[column].count(),
            'mean': df[column].mean(),
            'median': df[column].median(),
            'mode': mode_value,
            'variance': df[column].var(),
            'std_dev': df[column].std(),
            'min': df[column].min(),
            'max': df[column].max(),
            'range': df[column].max() - df[column].min(),
            '25%': df[column].quantile(0.25),
            '50% (median)': df[column].quantile(0.50),
            '75%': df[column].quantile(0.75),  # Corrected this line
        }

    # Return the stats as a DataFrame
    return pd.DataFrame(stats_summary)

# Call function and print result
descriptive_stats = descriptive_statistics(df)
print(descriptive_stats)
```
